```python
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format
```

```python
summary_billing_df = pd.read_csv("SUMMARY_BILLING_2023.csv")
```

```python
summary_billing_df.dtypes
```

```python
summary_billing_df.head()
```

```python
# Group by 'PRODUCT DESC' and sum 'AMOUNT_DUE'
product_due_sum = summary_billing_df.groupby('PRODUCT DESC')['AMOUNT_DUE'].sum().sort_values(ascending=False)
product_due_sum
```

```python
# List of desired product descriptions
desired_products = ['Fiber 2U', 'IP Phone']
# Filter rows based on 'PRODUCT DESC' using boolean indexing
filtered_df = summary_billing_df[summary_billing_df['PRODUCT DESC'].isin(desired_products)]
```

```python
filtered_df
```

```python
filtered_df.groupby(['ACCOUNT_TYPE', 'PRODUCT DESC'])['AMOUNT_DUE'].sum().sort_values(ascending=False)
```

```python
filtered_df.to_csv("Customer Group and Product Amount.csv", index=False, float_format="%.2f")
```
